```python
import numpy as np

fheader = "headernames.data"
fdata = "cleveland.data"

with open(fheader, 'r', encoding='ascii') as f:
    headers = [word for line in f for word in line.split()]

len(headers)
```

```python
import pandas as pd

with open(fdata, 'r', encoding='ascii') as f:
    values = [word for line in f for word in line.split()]
    data_processed = np.reshape(np.array(values), (-1, 76))

fuck = pd.DataFrame.from_records(data_processed, columns=np.array(headers))
fuck
```

```python
# fuck names
del fuck['name']
```

```python
# convert to numbers
fuck = fuck.applymap(float)
# responses
Y = fuck['num'].map(lambda x: 1 if x != 0 else 0)

# delete "garbage columns" that only have one value for the healthy
# as they probably just describe the response
for k in fuck.keys():
    if fuck[k][Y == 0].nunique() == 1:
        print('fuck ' + k)
        del fuck[k]
```

```python
# break down categorical variables

factors = ['cp', 'restecg', 'proto', 'slope', 'restwm']

for factor in factors:
    if factor in fuck:
        for value in fuck[factor].unique():
            fuck[factor + str(int(value))] = (fuck[factor] == value).map(flatten)
        del fuck[factor]
    
```

```python
problematic_cols = [k for k in fuck.keys() if any(fuck[k] == -9)]
print('Formerly problematic:', problematic_cols)
# fill in invalid data
for k in fuck.keys():
    k_bar = fuck[k].mean()
    fuck[k] = fuck[k].map(lambda x: k_bar if x == -9 else x)
    
problematic_cols = [k for k in fuck.keys() if any(fuck[k] == -9)]
print('Now problematic:', problematic_cols)
```
